```python
import re
def removeURLFromText(text):
    print(str(text))
    text = re.sub(r'b"*', '', str(text), flags=re.MULTILINE)
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', str(text), flags=re.MULTILINE)
    return (text)
```

```python
df['text'] = df['text'].apply(lambda x: removeURLFromText(x))
```
